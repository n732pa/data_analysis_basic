```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

qoute_wise = []
quote_author = []
for page in range(1,3):
    url='https://quotes.toscrape.com/tag/humor/page/' + str(page) + '/'
    html = requests.get(url).text
    soup = BeautifulSoup(html)

    # quote = soup.find_all('div',class_='quote')     #find_all -> 리스트로 생성
    # for i in range(len(quote)):
        # wise = quote[i].find('span',class_='text')
        # print(wise.text)

    quote1 = soup.select('div.quote > span.text')     
    # for i in quote:
        # print(i.text)
        
    for i in quote1:
        qoute_wise.append(i.text)

    quote2 = soup.select('div.quote > span > small.author')
    for i in quote2:
        quote_author.append(i.text)
        
df = pd.DataFrame({'wise':qoute_wise, 'author':quote_author})

df.to_csv('qoute.csv')
```
