```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

data = {'이름':['가나다','마바사','ABC','DEF'],
        '점수':[50,60,80,65]}

df = pd.DataFrame(data)

# 판다스의 plot 이용한 차트
# 데이터프레임.plot()
df.plot(kind='bar', x='이름', rot=0, ylim=[0,100],color='y')
for i in list(df.index):
        rate= f'{(df.loc[i]['점수']/df['점수'].sum())*100:.1f}%'
        plt.text(x=i-0.1, y=df.loc[i]['점수']-8, s=str(df.loc[i]['점수'])+'점')
        plt.text(x=i-0.15, y=df.loc[i]['점수']/2, s=rate)

meanval = df['점수'].mean()
plt.axhline(y=meanval, color='r', linestyle='dashed')
plt.text(x=0-0.2,y=meanval+1,s=f'평균:{meanval:.1f}')

plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.title('학생별 시험 점수')

plt.savefig('graphic01.png', dpi=400)
print('파일이 저장되었습니다')
plt.show()
#plt.text(x=1,y=60,s='값')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

# subplot 작성, matplotlib 이용
data = {'year' : [2010,2011,2012,2013,2014],
        'sales' : [500,800,1200,900,1500],
        'expenses' : [300,400,600,500,800]
}

df2 =pd.DataFrame(data)

# 서브플롯 생성
fig, axes = plt.subplots(2,2, figsize=(10,9))

# 선그래프
axes[0,0].plot(df2['year'],df2['sales'])
axes[0,0].set_title('sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)
axes[0,0].set_xticks(df2['year'])       # 숫자 문자 전환


# 막대그래프
axes[0,1].bar(df2['year'],df2['expenses'])
axes[0,1].set_title('Expenses')

# 산점도 그래프
axes[1,0].scatter(df2['sales'],df2['expenses'])
axes[1,0].set_title('Sales/Expenses')

# 히스토그램 그래프
axes[1,1].hist(df2['sales'],bins=3)             # bins = 구간개수
axes[1,1].set_title('Sale Distribution')
 
# 그래프 간 간격 조정
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')

# subplot 작성, seaborn 이용
data = {'year' : [2010,2011,2012,2013,2014],
        'sales' : [500,800,1200,900,1500],
        'expenses' : [300,400,600,500,800]
}

df2 =pd.DataFrame(data)

# 서브플롯 생성
fig, axes = plt.subplots(2,2, figsize=(10,9))

# 선 그래프
sns.lineplot(data=df2,x='year', y='sales', ax=axes[0,0])
axes[0,0].set_title('sales', fontsize=12)
axes[0,0].set_xlabel('year', fontsize=9)

# 막대 그래프
sns.barplot(data=df2,x='year', y='expenses', ax=axes[0,1], palette='Set1')

# 산점도 그래프
sns.scatterplot(data=df2,x='sales', y='expenses', ax=axes[1,0])

# 히스토그램 그래프
sns.histplot(data=df2,x='sales', ax=axes[1,1])

plt.show()
```
